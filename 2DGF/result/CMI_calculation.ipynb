```python
import numpy as np
label = ["-1.2", "-1.0", "-0.8","-0.5","0.0","0.5","0.8","1.2"]
```

```python
data = [[] for i in range (len(label))]
first_var = [[] for i in range (len(label))]
second_var = [[] for i in range (len(label))]
for i in range (len(label)):
    for j in range (40):
        if (i==len(label)-1):
            data[i].append(np.load(label[i]+"/tv_"+label[i]+"y9x4_param"+str(40*i+40+j+1)+".npy"))
            first_var[i].append(np.load(label[i]+"/tv_first_var"+label[i]+"y9x4_param"+str(40*i+40+j+1)+".npy"))
            second_var[i].append(np.load(label[i]+"/tv_second_var"+label[i]+"y9x4_param"+str(40*i+40+j+1)+".npy"))
        else:
            data[i].append(np.load(label[i]+"/tv_"+label[i]+"y9x4_param"+str(40*i+j+1)+".npy"))
            first_var[i].append(np.load(label[i]+"/tv_first_var"+label[i]+"y9x4_param"+str(40*i+j+1)+".npy"))
            second_var[i].append(np.load(label[i]+"/tv_second_var"+label[i]+"y9x4_param"+str(40*i+j+1)+".npy"))
```

```python
data=np.array(data)
first_var = np.array(first_var)
second_var = np.array(second_var)
print(data.shape)
```

```python
data_mean = np.mean(data, axis = (1,2))
first_var_mean = np.mean(first_var, axis=(1,2))
second_var_mean = np.mean(second_var, axis=(1,2))
```

```python
print(data_mean[2])
```

```python
data = np.mean(data, axis=0)
first_var = np.mean(first_var, axis=0)
second_var = np.mean(second_var, axis=0)
```

```python
import matplotlib.pyplot as plt
for i in range (len(label)):
    plt.plot(np.arange(13), (data_mean[i]), label="Jz="+label[i], alpha=0.4)
plt.xlabel("distance")
plt.ylabel("TV")
plt.title("2DXXZ_CMI_decay")
plt.legend(loc="lower right")

plt.savefig("2DXXZ_CMI.png", dpi = 150)
```
