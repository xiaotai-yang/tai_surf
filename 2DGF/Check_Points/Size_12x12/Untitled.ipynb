```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
XmeanE=np.load("meanEnergy_2DTGCRNN_X_12x12_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
ZmeanE=np.load("meanEnergy_2DTGCRNN_Z_12x12_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
XvarE=np.load("varEnergy_2DTGCRNN_X_12x12_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
ZvarE=np.load("varEnergy_2DTGCRNN_Z_12x12_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
```

```python
plt.plot(np.arange(XmeanE.shape[0]), XmeanE, label="X_measurement(No CMI decay)")
plt.plot(np.arange(ZmeanE.shape[0]), ZmeanE, label="Z_measurement(CMI decay)")
plt.ylim(-144,- 60)
plt.xlabel("training steps")
plt.ylabel("Mean_E")
plt.legend()
plt.savefig("GS_meanE.png", dpi=150)
```

```python
plt.plot(np.arange(XvarE.shape[0]), XvarE, label="X_measurement")
plt.plot(np.arange(ZvarE.shape[0]), ZvarE, label="Z_measurement")
plt.ylim(0,100)
plt.xlabel("training_step")
plt.ylabel("Var_E")
plt.legend()
plt.savefig("GS_Var_E.png", dpi=150)
```
