```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
XmeanE=np.load("meanEnergy_2DTGCRNN_X_10x10_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
ZmeanE=np.load("meanEnergy_2DTGCRNN_Z_10x10_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
XvarE=np.load("varEnergy_2DTGCRNN_X_10x10_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
ZvarE=np.load("varEnergy_2DTGCRNN_Z_10x10_units_128_numsamples_256_magfixedFalse_mag0_symmetry_nosym_num_gpus2.npy")
```

```python
plt.plot(np.arange(XmeanE.shape[0]), XmeanE, label="X measurement(No CMI decay)")
plt.plot(np.arange(ZmeanE.shape[0]), ZmeanE, label="Z measurement(CMI decay)")
plt.ylim(-100,- 60)
plt.xlabel("training steps", loc="right", fontsize = 16)
plt.ylabel("Energy", rotation = 'horizontal', loc="top",fontsize = 16)
plt.rcParams.update({'font.size': 14})
plt.legend(loc = "upper right")
plt.savefig("GS_meanE.png", dpi=150, bbox_inches='tight')
```

```python
plt.plot(np.arange(XvarE.shape[0]), XvarE, label="X_measurement")
plt.plot(np.arange(ZvarE.shape[0]), ZvarE, label="Z_measurement")
plt.ylim(0,150)
plt.xlabel("training_step")
plt.ylabel("Var_E")
plt.legend()
plt.savefig("GS_Var_E.png", dpi=150)
```
