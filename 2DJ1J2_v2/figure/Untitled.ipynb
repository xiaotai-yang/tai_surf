```python
import numpy as np
import matplotlib.pyplot as plt

J = ["0.0", "0.2", "0.4", "0.45", "0.5","0.55","0.6"]
log_correlation = [[] for i in range(7)]
for i in range (len(log_correlation)):
    log_correlation[i].append(np.load("log_correlation_y9x4_J="+J[i]+".npy"))
```

```python
for i in range(7):
    plt.plot(np.arange(15),  log_correlation[i][0][:], label=J[i],alpha=0.4)
plt.legend()
```
