```python
import numpy as np
```

```python
data = np.array([])
first_var = np.array([])
second_var = np.array([])
for i in range(160):
    data = np.append(data, np.load("0.0/tv_0.0y9x4_param"+str(i+1)+".npy"))
    first_var = np.append(first_var, np.load("0.0/tv_first_var0.0y9x4_param"+str(i+1)+".npy"))
    second_var = np.append(second_var, np.load("0.0/tv_second_var0.0y9x4_param"+str(i+1)+".npy"))
```

```python
data = np.reshape(data,(160, 13))
print(data)
first_var = np.reshape(first_var,(160, 13))
second_var = np.reshape(second_var, (160, 13))
```

```python
data = np.mean(data, axis=0)
first_var = np.mean(first_var, axis=0)
second_var = np.mean(second_var, axis=0)
```

```python
print(first_var)
print(second_var)
```

```python
import matplotlib.pyplot as plt
plt.scatter(np.arange(13), data)
plt.show()
```
