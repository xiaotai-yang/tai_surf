```python
import numpy as np
import matplotlib.pyplot as plt

J = ["0.5","1.0", "1.5", "2.0", "3.0", "3.5"]
log_correlation = [[] for i in range(len(J))]
for i in range (len(log_correlation)):
    log_correlation[i].append(np.load("log_connected_correlation_y9x4_J="+J[i]+".npy"))
```

```python
for i in range(len(J)):
    plt.plot(np.arange(17),  log_correlation[i][0][:], label="Bx="+J[i],alpha=0.4)
plt.xlabel("distance")
plt.ylabel("log_connected_correlation")
plt.title("2DTFIM_log_connected_correlation")
plt.legend()
plt.savefig("2DTFIM_log_connected_correlation")
```
