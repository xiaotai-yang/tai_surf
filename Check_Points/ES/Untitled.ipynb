```python
import numpy as np
Zmean = np.load("meanEnergy_N64_samp128_lradap0.0005_complexGRURNN_ES_units_64_zeromag.npy")
Zvar = np.load("varEnergy_N64_samp128_lradap0.0005_complexGRURNN_ES_units_64_zeromag.npy")
```

```python

import numpy as np
import ast

# Replace 'your_file.txt' with the actual path to your text file
with open('X_meanEnergy_N64_samp128_lradap0.0005_complexGRURNN_ES_units_64_zeromag.txt', 'r') as file:
    data_Xmean = [ast.literal_eval(line.strip()) for line in file]
with open('X_varEnergy_N64_samp128_lradap0.0005_complexGRURNN_ES_units_64_zeromag.txt', 'r') as file:
    data_Xvar = [ast.literal_eval(line.strip()) for line in file]
Xmean = np.array(data_Xmean)
Xvar = np.array(data_Xvar)

print(Xmean)
print(Xvar)
```

```python
Zmean = np.real(Zmean)
Xmean = np.real(Xmean)
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(len(Zmean)), Zmean, label = "Z measurment(No CMI decay)")
plt.plot(np.arange(len(Xmean)), Xmean, label = "X measurement(CMI decay)")

plt.rcParams.update({'font.size': 14})
plt.xlabel("training steps", loc="right", fontsize = 16)
plt.ylabel("Energy", rotation = 0, loc="top",fontsize = 16)
plt.ylim(top=-40) #ymax is your value
plt.ylim(bottom=-64) #ymin is your value
plt.legend()
plt.savefig("meanE_es.png", dpi =150, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(len(Zmean)), Zvar, label = "CMI_non_decay_case")
plt.plot(np.arange(len(Xmean)), Xvar, label = "CMI_decay_case")
plt.legend()
plt.ylabel("variance_E")
plt.xlabel("training_steps")
plt.ylim(top=50) #ymax is your value
plt.ylim(bottom=0) #ymin is your value
plt.savefig("varE_es.png", dpi =150)
```
